## Problema 3



```python
def adj(X):
    return det(X)*inv(X)
```

```python
a=[
    [1,-1,1],
    [0,2,-1],
    [1,0,0]
]

adj(a)
```

```python
b=[
    [2,5,8],
    [3,10,1],
    [1,15,4]
]

adj(b).round(3) #Redondear a 3 cifras despu√©s del punto
```

```python
c=[
    [5,2,-4],
    [6,-3,7],
    [1,2,4]
]

adj(c).round(3)
```

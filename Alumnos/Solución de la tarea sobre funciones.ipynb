¿Cómo hacer una tablita de números?

| x | y = f(x)|
|:----:|:----:|
| ... | ... |
| -1  | y=f(-1) |
| 0  | y=f(0) |
| 1  | y=f(1) |
| ... | ... |

Primer método: utilización de _lambdas_.
Queremos representar en la computadora esta función:
$$y=-3x-5$$
Lo que hacemos es:

```python
y=lambda x:-3*x-5
```

Esto nos permite evaluar $y$ dada una $x$. Por ejemplo:
$$y(-1)=-3(-1)-5=3-5=-2$$
En la compu sería:

```python
y(-1)
```

Análogamente:
$$y(0)=-3(0)-5=-5$$

```python
y(0)
```

Este procedimiento de evaluar `y` en distintas `x` se puede hacer de forma iterativa:

```python
X=[-2,-1,0,1,2]
Y=[y(x) for x in X]
print(f"{'x':<5} | {'y':>5}")
for a,b in zip(X,Y):
    print(f'{a:<5} | {b:>5}')
```

Del lado izquierdo aparecen los números del listado `[-2,-1,0,1,2]`, y del lado derecho salen los resultados de evaluar la expresión $y=-3x-5$ en cada uno de aquéllos números.
Esto lo podemos representar más simple:

```python
Y
```

> **Ejemplo**: Hacer una tabla de $f(x)=x^2-2$ para $x\in\{-3,-2,-1,0,1,2,3\}$

```python
#Definimos la lambda que representa a f
e=lambda a:a**2-2
```

```python
#Definimos el arreglito (listado) de números x para los que obtendremos y.
X=[-3,-2,-1,0,1,2,3]
Y=[e(c) for c in X]

print(f"{'x':<5} | {'y':>5}")
for p,q in zip(X,Y):
    print(f'{p:<5} | {q:>5}')
```

> **Ejemplo**: Hacer una tabla de $f(x)=x^6-5x^4+3x^2-7x+5$ para $x\in\{i\in\mathbb{N}: 0\leq i \leq 100\}$

```python
#numeros naturales entre 0 y 100
X=[j for j in range(101)]
f=lambda x:x**6-5*x**4+3*x**2-7*x+5
Y=[f(x) for x in X]
print(Y)
```

```python
#En una tablita
print(f"{'x':<5} | {'y':>5}")
print('-'*15)
for p,q in zip(X,Y):
    print(f'{p:<5} | {q:>5}')
```

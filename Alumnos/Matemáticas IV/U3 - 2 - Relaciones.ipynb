# Crear la gráfica de una relación

```python
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_edges_from(
    [
        ('a','b'),
        ('b','b'),('b','c'),('b','x'),
        ('d','m'),('d','k'),
        ('h','m'),('h','h'),
        ('k','k'),('x','b')
    ]
)
```

```python
%matplotlib inline

pos =nx.kamada_kawai_layout(G)
nx.draw_networkx(
    G,
    node_color='crimson',
    node_size=600,
    font_color='white',
    font_size=14,
    pos=pos
)
```

```python
H = nx.DiGraph(
    {
        'a':['a','p','q'],
        'b':['b','n'],
        's':['s'],
        'y':['y'],
        'c':['c','q'],
        'd':['n'],
        'm':['m','s'],
        'n':['n','b','c'],
        'p':['p'],
        'q':['q'],
        'r':['r','p']
    }
)
```

```python
%matplotlib inline

pos =nx.shell_layout(H)
nx.draw_networkx(
    H,
    node_color='crimson',
    node_size=400,
    font_color='white',
    font_size=12,
    pos=pos
)
```

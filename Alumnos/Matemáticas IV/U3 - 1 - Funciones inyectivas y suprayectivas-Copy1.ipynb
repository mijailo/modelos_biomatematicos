# Rutina para calcular el conjunto potencia

```python
def powerset(seq):
    """
    Función generadora de todos los subconjuntos de una lista.
    type(seq)=list
    """
    if len(seq) <= 1:
        yield seq
        yield []
    else:
        for item in powerset(seq[1:]):
            yield [seq[0]]+item
            yield item
```

# Rutina para calcular las funciones de A en B

```python
def funciones(A,B):
    AxB=[(i,j) for i in A for j in B]
    RelAB=powerset(AxB)
    j=0
    F_AB=[]
    for i in RelAB:
        if len(i)==len(A) and {k[0] for k in i}==set(A):
            F_AB.append(set(i))
    return F_AB
```

# Rutina para calcular las funciones *inyectivas* de A en B

```python
# Rutina para calcular las funciones *inyectivas* de A en B

def inyectivas(A,B):
    In_AB=[f for f in funciones(A,B) if len(set(x[1] for x in f))==len(A)]
    return In_AB
```

# Rutina para calcular las funciones *suprayectivas* de A en B

```python
def suprayectivas(A,B):
    Sup_AB=[f for f in funciones(A,B) if len(set(x[1] for x in f))==len(B)]
    return Sup_AB
```

# <font color='blue'> Unidad 2. Expresiones algebraicas para describir y generalizar </font>

## <font color='purple'> Objetivo específico </font>

El alumno desarrollará habilidades de razonamiento lógico, abstracción, generalización y comunicación matemática al: 
* Representar fenómenos o eventos a través de modelos algebraicos que involucren operaciones con expresiones algebraicas;
* Analizar representaciones y resolver las operaciones algebraicas involucradas mediante procedimientos diversos aplicando las propiedades pertinentes;
* Fundamentar el procedimiento algebraico seleccionado; 
* Validar sus resultados en el contexto de la situación o fenómeno analizado.

## <font color='purple'> Contenidos conceptuales </font>

1. Expresiones algebraicas para describir y generalizar patrones y relaciones numéricas en problemas naturales y sociales.
2. Expresiones algebraicas.
    1. Polinomiales 
    2. No polinomiales 
3. Propiedades para operar con expresiones numéricas y algebraicas radicales.

## <font color='purple'> Contenidos procedimentales </font>

4. Traducción de ideas entre el lenguaje materno y el lenguaje algebraico.
5. Operaciones con expresiones polinomiales y no polinomiales.
    1. Adición y sustracción
    2. Multiplicación y división
    3. División sintética
    4. Productos notables
    5. Factorización
    6. Expresiones con radicales (aritméticas y algebraicas)
    7. Simplificación de expresiones algebraicas racionales y con radicales
6. Modelación de situaciones que requieren el empleo de expresiones algebraicas, interpretación de los resultados y argumentación de sus conclusiones.

## <font color='purple'> Contenidos actitudinales </font>

7. Valoración de la importancia de la comunicación matemática para promover el desarrollo de su pensamiento abstracto al expresar sus ideas mediante el lenguaje propio, algebraico y gráfico.

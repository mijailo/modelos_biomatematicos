```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
x=np.geomspace(0.7,100,1000)
```

Definir $f(x)=bx^a$ (con $a,b>0$).

$\log_a x = y $

```python
def f(x0,a,b):
    return b*x0**a
```

```python
y0=[f(x0,1.83,2) for x0 in x]
y1=[f(x0,1,2) for x0 in x]
y2=[f(x0,0.55,2) for x0 in x]
```

```python
fig, ax = plt.subplots(
    1,1,
    figsize=(4,4)
)

ax.plot(x,y0, color='blue', lw=1.5, ls='-', label=r'$\alpha$')
ax.plot(x,y1, color='crimson', lw=1.5, ls=':', label=r'$\beta$')
ax.plot(x,y2, color='indigo', lw=1.5, ls='-.', label=r'$\gamma$')

ax.set_xscale('log')
ax.set_yscale('log')

plt.legend(fontsize=13)
plt.grid(
    which='both',
    alpha=0.5,
    ls=':',
    lw=0.5
)
plt.show()
```

Definir $g(x)=be^{ax}$ (con $a,b>0$).

```python
def g(x0,a,b):
    return b*np.exp(a*x0)
```

```python
z0=[g(x0,1.83,2) for x0 in x]
z1=[g(x0,1,2) for x0 in x]
z2=[g(x0,0.55,2) for x0 in x]
```

```python
fig, ax = plt.subplots(
    1,1,
    figsize=(4,4)
)

ax.plot(x,z0, color='blue', lw=2, ls='-', label=r'$\alpha$')
ax.plot(x,z1, color='crimson', lw=1.5, ls=':', label=r'$\beta$')
ax.plot(x,z2, color='indigo', lw=1.5, ls='-.', label=r'$\gamma$')

ax.set_xscale('linear')
ax.set_yscale('log')

plt.legend(fontsize=13)
plt.show()
```
